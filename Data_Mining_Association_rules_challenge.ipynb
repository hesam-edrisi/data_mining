```python
values = []
column = 'Age'
for row in df[column].values:
    val = [0,0,0,0]
    if 16 < row < 28:
        val[0]=1
    elif 27 < row < 36:
        val[1]=1
    elif 35 < row < 48:
        val[2]=1
    else:
        val[3]=1
    values.append(val)
values 
```

```python
new_colums = [f"{column}_gp{col}" for col in range(4)]
new_colums
```

```python
df[['Age_gp0', 'Age_gp1', 'Age_gp2', 'Age_gp3']] = values
df
```

```python
to_drop =['Elapsed_Time','Age']
df.drop(columns=to_drop, inplace=True)
df
```

```python
values = []
for row in df['Working'].values:
    val = [0]
    if  row == 'Yes':
        val[0]=1
    else:
        val[0]=0
    values.append(val)
values
```
